## matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2.Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3.Plot the data
ax.plot(x, y)

# 4.Customize plot
ax.set(title="Simple plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5.Save and show (save the entire figure)
fig.savefig("images/sample-plot.png")
```

```python
# Make a bar from dictionary
nut_butter_prices = {"Almond butter":10,
                     "Peanut butter":20,
                     "Cashew butter":30}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Butter prices",
       xlabel="butter type",
       ylabel="price");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
```

```python
#make some data for histogram and plot it.
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```

## Two options for subplots

```python
# subplot option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x, x/2)
ax2.hist(x)
ax3.scatter(np.random.random(10),np.random.random(10))
ax4.bar(nut_butter_prices.keys(),nut_butter_prices.values())
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].hist(x)
ax[1,1].bar(nut_butter_prices.keys(),nut_butter_prices.values())
```

# Matplotlib OO method to data visualization is recomended to use for advaced visualizations.
# Other than using the pyplot method. pyplot method is suitable for small visualizations but not for advanced things

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
over_50
```

```python
len(over_50)
```

```python
# pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
#ax.set_xlim(45,100);  aet x axis range from 45 to 100
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

#Customize the plot
ax.set(title="Heart Disease and Cholestrol Level",
       xlabel="Age",
       ylabel="Cholestrol")

#Add a legend
ax.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line over the mean value of the cholestrol
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
# subplot of chol, age and thelach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

#Customizing
ax0.set(title="Heart Disease and Cholestrol Level",
       #xlabel="Age",
       ylabel="Cholestrol")

#Add legend
ax0.legend(*scatter.legend_elements(),title="Target")

#Add horizontal line over the mean value of the cholestol
ax0.axhline(over_50["chol"].mean(),
           linestyle="--");

#Add data to ax1
scatter1 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"])

#customizing
ax1.set(title="Heart Disease and Thalach",
        xlabel="Age",
        ylabel="Thalach")

#Add legend
ax1.legend(*scatter1.legend_elements(),title="Target")

#Add horizontal line across the mean of the thalach
ax1.axhline(over_50["thalach"].mean(),
            linestyle="--")

#Title the figure
fig.suptitle("Heart Disease Analysis",
             fontsize="20",
             fontweight="bold");
```

```python
#Set the style
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter") #This is going to change the color

#Customize the plot
ax.set(title="Heart Disease and Cholestrol Level",
       xlabel="Age",
       ylabel="Cholestrol")

#Add a legend
ax.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line over the mean value of the cholestrol
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
# Cutomizing the x and y axis limitations

# subplot of chol, age and thelach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)



#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                     cmap="winter")

#Customizing
ax0.set(title="Heart Disease and Cholestrol Level",
       #xlabel="Age",
       ylabel="Cholestrol")

#Change the x-axis limit
ax0.set_xlim([50,80])

#Change the y axis limit
ax0.set_ylim([100,600])

#Add legend
ax0.legend(*scatter.legend_elements(),title="Target")

#Add horizontal line over the mean value of the cholestol
ax0.axhline(over_50["chol"].mean(),
           linestyle="--");

#Add data to ax1
scatter1 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"],
                       cmap="winter")

#customizing
ax1.set(title="Heart Disease and Thalach",
        xlabel="Age",
        ylabel="Thalach")

#Change the x-axis limit
ax1.set_xlim([50,80]) #Completely optional because scatters share the same x axis

#Change y axis limit not necessary to chage x limits because both scatters share the same x axis
ax1.set_ylim([60,200])

#Add legend
ax1.legend(*scatter1.legend_elements(),title="Target")

#Add horizontal line across the mean of the thalach
ax1.axhline(over_50["thalach"].mean(),
            linestyle="--")

#Title the figure
fig.suptitle("Heart Disease Analysis",
             fontsize="20",
             fontweight="bold");
```
